# 1.3 Select Features for RNN


```python
# Use CPI Index as the feature
data = df[['Index']].values

print(f"Data shape: {data.shape}")
```

# 1.4 Normalize the Data


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

print(f"Scaled data range: {scaled_data.min():.2f} to {scaled_data.max():.2f}")
```

# 1.5 Create Sliding Window Sequences


```python
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

# Define window size (e.g., 12 months = 1 year of history)
WINDOW_SIZE = 12

X, y = create_sequences(scaled_data, WINDOW_SIZE)

# Reshape X for RNN: (samples, timesteps, features)
X = X.reshape(X.shape[0], X.shape[1], 1)

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

# 1.6 Train-Test Split


```python
# Split into train (80%) and test (20%)
train_size = int(len(X) * 0.8)

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")
```

# 1.7 Save Preprocessed Data

#2.4 Stationarity Test (Augmented Dickey-Fuller)


```python
print("\n" + "="*60)
print("STATIONARITY TEST (Augmented Dickey-Fuller)")
print("="*60)

def adf_test(series, title):
    print(f"\n{title}:")
    result = adfuller(series.dropna(), autolag='AIC')
    print(f'ADF Statistic: {result[0]:.6f}')
    print(f'p-value: {result[1]:.6f}')
    print(f'Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value:.3f}')

    if result[1] <= 0.05:
        print("✓ Series is STATIONARY (reject H0)")
    else:
        print("✗ Series is NON-STATIONARY (fail to reject H0)")

# Test CPI Index (likely non-stationary)
adf_test(df['Index'], "CPI Index")

# Test Inflation Rate (likely stationary)
adf_test(df['Inflation'], "Inflation Rate")
```

# Step 3: Build Baseline RNN Model


```python
# STEP 1 RECOVERY: Create preprocessed data files
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import joblib
from google.colab import drive

# Mount Drive
drive.mount('/content/drive')

# Create RNN_Project folder if it doesn't exist
import os
os.makedirs('/content/drive/MyDrive/RNN_Project', exist_ok=True)

# Load data
file_path = '/content/drive/MyDrive/RNN_Project/cpiai_csv.csv'

# Check if file exists in the project folder
if not os.path.exists(file_path):
    # If not, try to find it in the root Drive
    alt_path = '/content/drive/MyDrive/cpiai_csv.csv'
    if os.path.exists(alt_path):
        file_path = alt_path
        print(f"Found CSV at: {file_path}")
    else:
        print("Please upload the CSV file first using the code below:")
        from google.colab import files
        uploaded = files.upload()
        file_path = 'cpiai_csv.csv'

df = pd.read_csv(file_path)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.dropna().copy()

print(f"✓ Data loaded: {df.shape[0]} rows")

# Use CPI Index as feature
data = df[['Index']].values

# Normalize
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Create sequences
def create_sequences(data, window_size=12):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

WINDOW_SIZE = 12
X, y = create_sequences(scaled_data, WINDOW_SIZE)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Train-test split
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Save to Drive
np.save('/content/drive/MyDrive/RNN_Project/X_train.npy', X_train)
np.save('/content/drive/MyDrive/RNN_Project/X_test.npy', X_test)
np.save('/content/drive/MyDrive/RNN_Project/y_train.npy', y_train)
np.save('/content/drive/MyDrive/RNN_Project/y_test.npy', y_test)
joblib.dump(scaler, '/content/drive/MyDrive/RNN_Project/scaler.pkl')

print("\n" + "="*50)
print("✓ Step 1 Recovery Complete!")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"Files saved to: /content/drive/MyDrive/RNN_Project/")
print("="*50)
```

#3.1 Load Preprocessed Data from Step 1


```python
# Load the preprocessed data saved in Step 1
X_train = np.load('/content/drive/MyDrive/RNN_Project/X_train.npy')
X_test = np.load('/content/drive/MyDrive/RNN_Project/X_test.npy')
y_train = np.load('/content/drive/MyDrive/RNN_Project/y_train.npy')
y_test = np.load('/content/drive/MyDrive/RNN_Project/y_test.npy')
scaler = joblib.load('/content/drive/MyDrive/RNN_Project/scaler.pkl')

print(f"Training data shape: X_train={X_train.shape}, y_train={y_train.shape}")
print(f"Test data shape: X_test={X_test.shape}, y_test={y_test.shape}")
print(f"Input shape for RNN: (samples, timesteps, features) = {X_train.shape}")
```

#3.2 Build Simple RNN Model


```python
# Define the RNN architecture
model = Sequential([
    # SimpleRNN layer with 50 units
    SimpleRNN(50, activation='tanh', input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False),

    # Dropout for regularization (prevents overfitting)
    Dropout(0.2),

    # Output layer (single value prediction)
    Dense(1)
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='mse',  # Mean Squared Error
    metrics=['mae', 'mse']  # Mean Absolute Error
)

# Display model architecture
print("\n" + "="*60)
print("MODEL ARCHITECTURE")
print("="*60)
model.summary()
```

#3.3 Train the Model


```python
# Callbacks for better training
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    ModelCheckpoint(
        '/content/drive/MyDrive/RNN_Project/best_rnn_model.h5',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    )
]

# Train the model
print("\n" + "="*60)
print("TRAINING THE RNN MODEL")
print("="*60)

history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_split=0.2,  # Use 20% of training for validation
    callbacks=callbacks,
    verbose=1
)

# Save the final model
model.save('/content/drive/MyDrive/RNN_Project/rnn_model_final.h5')
print("\n✓ Model saved to Drive")
```

#3.4 Visualize Training History


```python
# Plot training and validation loss
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Loss plot
axes[0].plot(history.history['loss'], label='Training Loss', linewidth=2)
axes[0].plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
axes[0].set_title('Model Loss During Training', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Loss (MSE)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# MAE plot
axes[1].plot(history.history['mae'], label='Training MAE', linewidth=2)
axes[1].plot(history.history['val_mae'], label='Validation MAE', linewidth=2)
axes[1].set_title('Model MAE During Training', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Mean Absolute Error')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/content/drive/MyDrive/RNN_Project/training_history.png', dpi=100)
plt.show()
```

#3.5 Make Predictions


```python
print("\n" + "="*60)
print("MAKING PREDICTIONS")
print("="*60)

# Predict on training and test data
y_train_pred = model.predict(X_train, verbose=0)
y_test_pred = model.predict(X_test, verbose=0)

# Inverse transform to get actual CPI values
y_train_actual = scaler.inverse_transform(y_train.reshape(-1, 1))
y_train_pred_actual = scaler.inverse_transform(y_train_pred)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
y_test_pred_actual = scaler.inverse_transform(y_test_pred)

print("✓ Predictions completed")
print(f"Training predictions shape: {y_train_pred.shape}")
print(f"Test predictions shape: {y_test_pred.shape}")
```

#3.6 Evaluate Model Performance


```python
# Calculate metrics
def calculate_metrics(y_actual, y_pred, dataset_name):
    mse = mean_squared_error(y_actual, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_actual, y_pred)
    r2 = r2_score(y_actual, y_pred)

    print(f"\n{dataset_name} SET METRICS:")
    print(f"  MSE:  {mse:.4f}")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE:  {mae:.4f}")
    print(f"  R²:   {r2:.4f}")

    return {'MSE': mse, 'RMSE': rmse, 'MAE': mae, 'R2': r2}

# Calculate metrics for both sets
train_metrics = calculate_metrics(y_train_actual, y_train_pred_actual, "TRAINING")
test_metrics = calculate_metrics(y_test_actual, y_test_pred_actual, "TEST")

# Save metrics to file
metrics_df = pd.DataFrame([train_metrics, test_metrics], index=['Training', 'Test'])
metrics_df.to_csv('/content/drive/MyDrive/RNN_Project/model_metrics.csv')
print("\n✓ Metrics saved to Drive")
```
