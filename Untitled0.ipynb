# 1.3 Select Features for RNN


```python
# Use CPI Index as the feature
data = df[['Index']].values

print(f"Data shape: {data.shape}")
```

# 1.4 Normalize the Data


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

print(f"Scaled data range: {scaled_data.min():.2f} to {scaled_data.max():.2f}")
```

# 1.5 Create Sliding Window Sequences


```python
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

# Define window size (e.g., 12 months = 1 year of history)
WINDOW_SIZE = 12

X, y = create_sequences(scaled_data, WINDOW_SIZE)

# Reshape X for RNN: (samples, timesteps, features)
X = X.reshape(X.shape[0], X.shape[1], 1)

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

# 1.6 Train-Test Split


```python
# Split into train (80%) and test (20%)
train_size = int(len(X) * 0.8)

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")
```

#2.4 Stationarity Test (Augmented Dickey-Fuller)


```python
print("\n" + "="*60)
print("STATIONARITY TEST (Augmented Dickey-Fuller)")
print("="*60)

def adf_test(series, title):
    print(f"\n{title}:")
    result = adfuller(series.dropna(), autolag='AIC')
    print(f'ADF Statistic: {result[0]:.6f}')
    print(f'p-value: {result[1]:.6f}')
    print(f'Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value:.3f}')

    if result[1] <= 0.05:
        print("✓ Series is STATIONARY (reject H0)")
    else:
        print("✗ Series is NON-STATIONARY (fail to reject H0)")

# Test CPI Index (likely non-stationary)
adf_test(df['Index'], "CPI Index")

# Test Inflation Rate (likely stationary)
adf_test(df['Inflation'], "Inflation Rate")
```

# Step 3: Build Baseline RNN Model


```python
# STEP 1 RECOVERY: Create preprocessed data files
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import joblib
from google.colab import drive

# Mount Drive
drive.mount('/content/drive')

# Create RNN_Project folder if it doesn't exist
import os
os.makedirs('/content/drive/MyDrive/RNN_Project', exist_ok=True)

# Load data
file_path = '/content/drive/MyDrive/RNN_Project/cpiai_csv.csv'

# Check if file exists in the project folder
if not os.path.exists(file_path):
    # If not, try to find it in the root Drive
    alt_path = '/content/drive/MyDrive/cpiai_csv.csv'
    if os.path.exists(alt_path):
        file_path = alt_path
        print(f"Found CSV at: {file_path}")
    else:
        print("Please upload the CSV file first using the code below:")
        from google.colab import files
        uploaded = files.upload()
        file_path = 'cpiai_csv.csv'

df = pd.read_csv(file_path)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.dropna().copy()

print(f"✓ Data loaded: {df.shape[0]} rows")

# Use CPI Index as feature
data = df[['Index']].values

# Normalize
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Create sequences
def create_sequences(data, window_size=12):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

WINDOW_SIZE = 12
X, y = create_sequences(scaled_data, WINDOW_SIZE)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Train-test split
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Save to Drive
np.save('/content/drive/MyDrive/RNN_Project/X_train.npy', X_train)
np.save('/content/drive/MyDrive/RNN_Project/X_test.npy', X_test)
np.save('/content/drive/MyDrive/RNN_Project/y_train.npy', y_train)
np.save('/content/drive/MyDrive/RNN_Project/y_test.npy', y_test)
joblib.dump(scaler, '/content/drive/MyDrive/RNN_Project/scaler.pkl')

print("\n" + "="*50)
print("✓ Step 1 Recovery Complete!")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"Files saved to: /content/drive/MyDrive/RNN_Project/")
print("="*50)
```

#3.1 Load Preprocessed Data from Step 1


```python
# Load the preprocessed data saved in Step 1
X_train = np.load('/content/drive/MyDrive/RNN_Project/X_train.npy')
X_test = np.load('/content/drive/MyDrive/RNN_Project/X_test.npy')
y_train = np.load('/content/drive/MyDrive/RNN_Project/y_train.npy')
y_test = np.load('/content/drive/MyDrive/RNN_Project/y_test.npy')
scaler = joblib.load('/content/drive/MyDrive/RNN_Project/scaler.pkl')

print(f"Training data shape: X_train={X_train.shape}, y_train={y_train.shape}")
print(f"Test data shape: X_test={X_test.shape}, y_test={y_test.shape}")
print(f"Input shape for RNN: (samples, timesteps, features) = {X_train.shape}")
```

#3.2 Build Simple RNN Model


```python
# Define the RNN architecture
model = Sequential([
    # SimpleRNN layer with 50 units
    SimpleRNN(50, activation='tanh', input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False),

    # Dropout for regularization (prevents overfitting)
    Dropout(0.2),

    # Output layer (single value prediction)
    Dense(1)
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='mse',  # Mean Squared Error
    metrics=['mae', 'mse']  # Mean Absolute Error
)

# Display model architecture
print("\n" + "="*60)
print("MODEL ARCHITECTURE")
print("="*60)
model.summary()
```

#3.3 Train the Model


```python
# Callbacks for better training
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    ModelCheckpoint(
        '/content/drive/MyDrive/RNN_Project/best_rnn_model.h5',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    )
]

# Train the model
print("\n" + "="*60)
print("TRAINING THE RNN MODEL")
print("="*60)

history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_split=0.2,  # Use 20% of training for validation
    callbacks=callbacks,
    verbose=1
)

# Save the final model
model.save('/content/drive/MyDrive/RNN_Project/rnn_model_final.h5')
print("\n✓ Model saved to Drive")
```

#3.5 Make Predictions


```python
print("\n" + "="*60)
print("MAKING PREDICTIONS")
print("="*60)

# Predict on training and test data
y_train_pred = model.predict(X_train, verbose=0)
y_test_pred = model.predict(X_test, verbose=0)

# Inverse transform to get actual CPI values
y_train_actual = scaler.inverse_transform(y_train.reshape(-1, 1))
y_train_pred_actual = scaler.inverse_transform(y_train_pred)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
y_test_pred_actual = scaler.inverse_transform(y_test_pred)

print("✓ Predictions completed")
print(f"Training predictions shape: {y_train_pred.shape}")
print(f"Test predictions shape: {y_test_pred.shape}")
```

#3.6 Evaluate Model Performance


```python
# Calculate metrics
def calculate_metrics(y_actual, y_pred, dataset_name):
    mse = mean_squared_error(y_actual, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_actual, y_pred)
    r2 = r2_score(y_actual, y_pred)

    print(f"\n{dataset_name} SET METRICS:")
    print(f"  MSE:  {mse:.4f}")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE:  {mae:.4f}")
    print(f"  R²:   {r2:.4f}")

    return {'MSE': mse, 'RMSE': rmse, 'MAE': mae, 'R2': r2}

# Calculate metrics for both sets
train_metrics = calculate_metrics(y_train_actual, y_train_pred_actual, "TRAINING")
test_metrics = calculate_metrics(y_test_actual, y_test_pred_actual, "TEST")

# Save metrics to file
metrics_df = pd.DataFrame([train_metrics, test_metrics], index=['Training', 'Test'])
metrics_df.to_csv('/content/drive/MyDrive/RNN_Project/model_metrics.csv')
print("\n✓ Metrics saved to Drive")
```

```python
# Load original data to create multi-feature dataset
df = pd.read_csv('/content/drive/MyDrive/RNN_Project/cpiai_csv.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.dropna().copy()

print("Original data shape:", df.shape)
print(df.head())

# Create multi-feature dataset (CPI Index + Inflation Rate)
from sklearn.preprocessing import MinMaxScaler

# Separate scalers for different features
scaler_cpi = MinMaxScaler(feature_range=(0, 1))
scaler_inflation = MinMaxScaler(feature_range=(0, 1))

# Scale both features
scaled_cpi = scaler_cpi.fit_transform(df[['Index']])
scaled_inflation = scaler_inflation.fit_transform(df[['Inflation']])

# Combine features
scaled_data_multi = np.column_stack((scaled_cpi, scaled_inflation))
print(f"Multi-feature data shape: {scaled_data_multi.shape}")

# Create sequences with multiple features
def create_sequences_multi(data, window_size=12):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, :])  # All features
        y.append(data[i, 0])  # Predict CPI Index
    return np.array(X), np.array(y)

WINDOW_SIZE = 12
X_multi, y_multi = create_sequences_multi(scaled_data_multi, WINDOW_SIZE)

# Train-test split
train_size = int(len(X_multi) * 0.8)
X_train_multi, X_test_multi = X_multi[:train_size], X_multi[train_size:]
y_train_multi, y_test_multi = y_multi[:train_size], y_multi[train_size:]

print(f"\nMulti-feature data shape:")
print(f"X_train: {X_train_multi.shape}")
print(f"X_test: {X_test_multi.shape}")
print(f"Features: CPI Index + Inflation Rate")

# Save multi-feature data for later use
np.save('/content/drive/MyDrive/RNN_Project/X_train_multi.npy', X_train_multi)
np.save('/content/drive/MyDrive/RNN_Project/X_test_multi.npy', X_test_multi)
np.save('/content/drive/MyDrive/RNN_Project/y_train_multi.npy', y_train_multi)
np.save('/content/drive/MyDrive/RNN_Project/y_test_multi.npy', y_test_multi)
joblib.dump(scaler_cpi, '/content/drive/MyDrive/RNN_Project/scaler_cpi.pkl')
joblib.dump(scaler_inflation, '/content/drive/MyDrive/RNN_Project/scaler_inflation.pkl')
```

###4.2 Compare Different RNN Architectures


```python
print("\n" + "="*60)
print("COMPARING DIFFERENT RNN ARCHITECTURES")
print("="*60)

def create_model(model_type='LSTM', units=50, dropout=0.2, learning_rate=0.001):
    """Create different RNN models for comparison"""
    model = Sequential()

    if model_type == 'LSTM':
        model.add(LSTM(units, activation='tanh', input_shape=(X_train_multi.shape[1], X_train_multi.shape[2])))
    elif model_type == 'GRU':
        model.add(GRU(units, activation='tanh', input_shape=(X_train_multi.shape[1], X_train_multi.shape[2])))
    elif model_type == 'SimpleRNN':
        model.add(SimpleRNN(units, activation='tanh', input_shape=(X_train_multi.shape[1], X_train_multi.shape[2])))

    model.add(Dropout(dropout))
    model.add(Dense(1))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

    return model

# Train and evaluate different architectures
architectures = ['LSTM', 'GRU', 'SimpleRNN']
results = {}
histories = {}

for arch in architectures:
    print(f"\n{'='*50}")
    print(f"Training {arch} Model...")
    print(f"{'='*50}")

    model = create_model(model_type=arch, units=50, dropout=0.2)

    callbacks = [
        EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=0),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=0)
    ]

    history = model.fit(
        X_train_multi, y_train_multi,
        epochs=80,
        batch_size=32,
        validation_split=0.2,
        callbacks=callbacks,
        verbose=0
    )

    # Evaluate
    y_pred = model.predict(X_test_multi, verbose=0)
    y_test_actual = scaler_cpi.inverse_transform(y_test_multi.reshape(-1, 1))
    y_pred_actual = scaler_cpi.inverse_transform(y_pred)

    rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred_actual))
    mae = mean_absolute_error(y_test_actual, y_pred_actual)
    r2 = r2_score(y_test_actual, y_pred_actual)

    results[arch] = {'RMSE': rmse, 'MAE': mae, 'R2': r2, 'model': model}
    histories[arch] = history

    print(f"\n{arch} Results:")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE:  {mae:.4f}")
    print(f"  R²:   {r2:.4f}")

# Compare results
print("\n" + "="*60)
print("ARCHITECTURE COMPARISON SUMMARY")
print("="*60)
comparison_df = pd.DataFrame(results).T
print(comparison_df.round(4))

# Find best model
best_arch = comparison_df['R2'].idxmax()
print(f"\n✓ Best architecture: {best_arch} (R² = {comparison_df.loc[best_arch, 'R2']:.4f})")

# Save comparison
comparison_df.to_csv('/content/drive/MyDrive/RNN_Project/architecture_comparison.csv')
```

###4.4 Hyperparameter Tuning for Best Architecture


```python
print("\n" + "="*60)
print(f"HYPERPARAMETER TUNING FOR {best_arch.upper()}")
print("="*60)

# Define hyperparameter grid
param_grid = {
    'units': [32, 64, 128],
    'dropout': [0.1, 0.2, 0.3],
    'learning_rate': [0.001, 0.0005, 0.0001],
    'layers': [1, 2]  # Number of RNN layers
}

def create_tuned_model(model_type, units, dropout, learning_rate, layers):
    """Create model with specified hyperparameters"""
    model = Sequential()

    # First layer
    if model_type == 'LSTM':
        model.add(LSTM(units, activation='tanh', return_sequences=(layers > 1),
                       input_shape=(X_train_multi.shape[1], X_train_multi.shape[2])))
    else:
        model.add(GRU(units, activation='tanh', return_sequences=(layers > 1),
                      input_shape=(X_train_multi.shape[1], X_train_multi.shape[2])))

    model.add(Dropout(dropout))

    # Additional layers
    for i in range(1, layers):
        if model_type == 'LSTM':
            model.add(LSTM(units // 2, activation='tanh', return_sequences=(i < layers-1)))
        else:
            model.add(GRU(units // 2, activation='tanh', return_sequences=(i < layers-1)))
        model.add(Dropout(dropout))

    model.add(Dense(1))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

    return model

# Perform grid search (limited combinations due to time)
print("\nPerforming grid search...")
best_score = -np.inf
best_params = {}
results_tuning = []

# Sample a subset of combinations for efficiency
combinations = [
    {'units': 64, 'dropout': 0.2, 'learning_rate': 0.001, 'layers': 1},
    {'units': 128, 'dropout': 0.2, 'learning_rate': 0.001, 'layers': 1},
    {'units': 64, 'dropout': 0.3, 'learning_rate': 0.001, 'layers': 1},
    {'units': 64, 'dropout': 0.2, 'learning_rate': 0.0005, 'layers': 1},
    {'units': 64, 'dropout': 0.2, 'learning_rate': 0.001, 'layers': 2},
    {'units': 32, 'dropout': 0.2, 'learning_rate': 0.001, 'layers': 1},
]

for i, params in enumerate(combinations, 1):
    print(f"\nTesting combination {i}/{len(combinations)}: {params}")

    model = create_tuned_model(
        best_arch,
        units=params['units'],
        dropout=params['dropout'],
        learning_rate=params['learning_rate'],
        layers=params['layers']
    )

    callbacks = [
        EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True, verbose=0),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, verbose=0)
    ]

    history = model.fit(
        X_train_multi, y_train_multi,
        epochs=60,
        batch_size=32,
        validation_split=0.2,
        callbacks=callbacks,
        verbose=0
    )

    # Evaluate
    y_pred = model.predict(X_test_multi, verbose=0)
    y_test_actual = scaler_cpi.inverse_transform(y_test_multi.reshape(-1, 1))
    y_pred_actual = scaler_cpi.inverse_transform(y_pred)

    rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred_actual))
    r2 = r2_score(y_test_actual, y_pred_actual)

    results_tuning.append({
        **params,
        'RMSE': rmse,
        'R2': r2,
        'model': model
    })

    print(f"  RMSE: {rmse:.4f}, R²: {r2:.4f}")

    if r2 > best_score:
        best_score = r2
        best_params = params
        best_model = model

print("\n" + "="*60)
print("BEST HYPERPARAMETERS FOUND")
print("="*60)
for key, value in best_params.items():
    print(f"  {key}: {value}")
print(f"  Best R² Score: {best_score:.4f}")

# Save best model
best_model.save('/content/drive/MyDrive/RNN_Project/best_tuned_model.h5')
print("\n✓ Best tuned model saved to Drive")
```

###5.5 Multi-Step Forecasting


```python
print("\n" + "="*60)
print("MULTI-STEP FORECASTING (Next 12 Months)")
print("="*60)

def forecast_future(model, last_sequence, scaler_cpi, scaler_inflation, steps=12):
    """
    Forecast future CPI values for specified number of steps
    """
    current_sequence = last_sequence.copy()
    forecasts = []

    for step in range(steps):
        # Predict next value
        next_pred_scaled = model.predict(current_sequence.reshape(1, current_sequence.shape[0], current_sequence.shape[1]), verbose=0)

        # Store prediction
        forecasts.append(next_pred_scaled[0, 0])

        # Update sequence for next prediction
        # For inflation, we'll use a simple persistence model (last known inflation)
        new_row = np.array([[next_pred_scaled[0, 0], current_sequence[-1, 1]]])
        current_sequence = np.vstack([current_sequence[1:], new_row])

    # Inverse transform forecasts
    forecasts_actual = scaler_cpi.inverse_transform(np.array(forecasts).reshape(-1, 1))

    return forecasts_actual.flatten()

# Get the last sequence from the data
last_sequence = X_test_multi[-1]  # Last test sequence
print(f"Last sequence shape: {last_sequence.shape}")

# Generate forecasts for next 12 months
forecast_steps = 12
forecasts = forecast_future(best_model, last_sequence, scaler_cpi, scaler_inflation, forecast_steps)

# Create forecast dates
last_date = df.index[-1]
forecast_dates = pd.date_range(start=last_date + pd.DateOffset(months=1), periods=forecast_steps, freq='M')

# Create forecast dataframe
forecast_df = pd.DataFrame({
    'Date': forecast_dates,
    'Forecasted_CPI': forecasts,
    'Lower_Bound': forecasts * 0.98,  # 2% lower confidence
    'Upper_Bound': forecasts * 1.02   # 2% upper confidence
})

print("\nForecasted CPI Values (Next 12 Months):")
print("-" * 60)
print(forecast_df.to_string(index=False))
print(f"\nAverage Forecasted CPI: {forecasts.mean():.2f}")
print(f"Expected Annual Growth: {((forecasts[-1] - df['Index'].iloc[-1]) / df['Index'].iloc[-1] * 100):.2f}%")

# Save forecast
forecast_df.to_csv('/content/drive/MyDrive/RNN_Project/forecast_next_12_months.csv', index=False)
```
